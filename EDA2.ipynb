### Handling the missing Values 

We will going to handle the missing values of those columns which have the missing values of 98 other than 1 and 2 .

```python

def replace_missing_values_percentage(column):
    # Calculate the percentage of occurrence for values 1 and 2
    value_counts = column.value_counts(normalize=True)
    
    if 98 in value_counts.index:
        percentage_1 = value_counts[1] if 1 in value_counts.index else 0
        percentage_2 = value_counts[2] if 2 in value_counts.index else 0
        
        # Normalize probabilities to ensure they sum to 1
        total_percentage = percentage_1 + percentage_2
        if total_percentage > 0:
            probability_1 = percentage_1 / total_percentage
            probability_2 = percentage_2 / total_percentage
            
            # Replace missing values (98) based on probabilities
            column = column.apply(lambda x: np.random.choice([1, 2], p=[probability_1, probability_2]) if x == 98 else x)
    
    return column

# Example usage for the TOBACCO column
dataset['TOBACCO'] = replace_missing_values_percentage(dataset['TOBACCO'])

# Print the updated TOBACCO column
print(f'TOBACCO: {dataset["TOBACCO"]}')

dataset['TOBACCO'].value_counts()




#now using for every column
dataset['DIABETES'] = replace_missing_values_percentage(dataset['DIABETES'])
dataset['HIPERTENSION'] = replace_missing_values_percentage(dataset['HIPERTENSION'])
dataset['COPD'] = replace_missing_values_percentage(dataset['COPD'])
dataset['CARDIOVASCULAR'] = replace_missing_values_percentage(dataset['CARDIOVASCULAR'])
dataset['RENAL_CHRONIC'] = replace_missing_values_percentage(dataset['RENAL_CHRONIC'])
dataset['OTHER_DISEASE'] = replace_missing_values_percentage(dataset['OTHER_DISEASE'])
dataset['INMSUPR'] = replace_missing_values_percentage(dataset['INMSUPR'])

dataset['INMSUPR'].value_counts()
dataset['DIABETES'].value_counts()
dataset['HIPERTENSION'].value_counts()
dataset['COPD'].value_counts()
dataset['CARDIOVASCULAR'].value_counts()
dataset['RENAL_CHRONIC'].value_counts()
dataset['OTHER_DISEASE'].value_counts()
dataset['TOBACCO'].value_counts()



#missing values handled .
```

## 4. SVC

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC


# Select a subset of data for initial experimentation and tuning
# You may adjust the sample size based on your computational resources and needs
subset_df = df2.sample(frac=0.1, random_state=72)


from sklearn.model_selection import train_test_split
X =  subset_df.drop('MORTALITY', axis = 1)
y = subset_df['MORTALITY']


# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.05, random_state=72)

# Apply PCA to reduce dimensionality
pca = PCA(n_components=0.95, random_state=42)  # Retain 95% of variance
X_train_pca = pca.fit_transform(X_train)
X_test_pca = pca.transform(X_test)

#Random under sampling
from imblearn.under_sampling import RandomUnderSampler
X_under, y_under = RandomUnderSampler().fit_resample(X_train_pca, y_train)
print(X_under.shape, y_under.shape)

# Perform feature scaling
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_under)
X_test_scaled = scaler.transform(X_test_pca)

# Initialize the Support Vector Classifier with a linear kernel
svm_clf = SVC(kernel='linear', random_state=72)

# Train the classifier
svm_clf.fit(X_train_scaled, y_under)

# Make predictions on the testing set
y_pred = svm_clf.predict(X_test_scaled)

# Calculate accuracy
accuracy = accuracy_score(y_test, y_pred)
```
